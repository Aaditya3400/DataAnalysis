### Data Cleaning

```python
data['country'].fillna(data['country'].mode(),inplace=True)
```

```python
data['children'].fillna(data['children'].mean(),inplace=True)
```

### Most of the values in company column are null so we can drop this column. 

```python
data.drop('company',axis=1,inplace=True)
```

### Deleting the invalid data (where adult = 0,children = 0 and babies = 0)

```python
delete = (data['adults']==0) & (data['children']==0) & (data['babies']==0)
```

```python
data = data[~delete]
```

# Exploratory Data Analysis

#  1. What type of hotel does a customer book in general.

```python
sns.countplot(data=data,x='hotel').set(xlabel='Hotel Type',ylabel='No. of Customer',title='Customer booking in general')
```

```python
dataHotel = data['hotel'].value_counts()
dataHotel.plot(kind='pie',explode=[0,.1],autopct='%1.1f%%')
plt.ylabel("")
plt.title("Booking type in percentage")
```

### Conclution: Customer likes to book city hotel in general. 66.4% peoples book city hotel and 33.6% peoples book resort hotel. 

#  2. Which year most of customers arrive ?

```python
df1 = data[data['is_canceled']==0]
```

```python
arrivedCustomer = df1.groupby('arrival_date_year')['hotel'].value_counts().unstack()
```

```python
arrivedCustomer['Total'] = arrivedCustomer['City Hotel'] + arrivedCustomer['Resort Hotel']
```

```python
arrivedCustomer.plot(kind='bar')
plt.xlabel("Arrival Year",fontsize=12)
plt.ylabel("No. of customer",fontsize=12)
plt.title("Most Customer Arrived Year",fontsize=15)
```

### Conclution: The maximun number of customer arrived in year 2016 in both type of hotels.

#  3. Which are the busiest months in the year of hotel booking ?

```python
busyMonth = data.groupby('arrival_date_month')['hotel'].value_counts().unstack()
```

```python
busyMonth.plot(kind='bar',figsize=(12,6))
plt.xlabel("Arrival Month",fontsize=14)
plt.ylabel("No. of customer",fontsize=14)
plt.title("Most Busiest Month",fontsize=15)
```

```python
busyMonth['total'] =busyMonth['City Hotel'] + busyMonth['Resort Hotel']
```

```python
busyMonth['total'].plot(kind='pie',explode=[.1,.25,.1,.1,.1,.25,.1,.1,.25,.1,.1,.1],autopct='%1.1f%%')
```

### Conclution: The month of August is most busiest month of the year in both type of hotels. July and may are the second and third busiest month respectively

#  4. Deposit type payment of customer ?

```python
df = data.groupby('deposit_type')['hotel'].value_counts().unstack()
```

```python
df.plot(kind='bar')
plt.xlabel("Deposite Type",fontsize=14)
plt.ylabel("No. of customer",fontsize=14)
plt.title("Customer Deposite Type",fontsize=15)
```

```python
df['Total'] = df['City Hotel'] + df['Resort Hotel']
```

```python
df.plot(kind='pie',y='Total',explode=[0,.1,.2],autopct='%1.1f%%',legend=False)
plt.ylabel('')
plt.title("Customer Deposite Type",fontsize=15)
```

## Conclution: 
#### No Deposit -    No deposit was made (around 87.6% bookings). 
#### Non Refund -  A deposit was made in the value of the total stay cost (around 12.2% bookings).
#### Refundable -   A deposit was made with a value under the cost of stay (around 0.1% bookings).

#  5. How many repeated customers?

```python
df = data.groupby('is_repeated_guest')['hotel'].value_counts().unstack()
```

```python
df.plot(kind='bar')
plt.xticks(ticks=[0,1],labels=['Not Repeated','Repeated'])
plt.xlabel("Repeated Customer",fontsize=14)
plt.ylabel("No. of customer",fontsize=14)
plt.title("Customer Type",fontsize=15)
```

```python
df['Total'] = df['City Hotel'] + df['Resort Hotel']
df['labels'] = ['Not Repeated','Repeated']
df.set_index('labels',inplace=True)
```

```python
df.plot(kind='pie',y='Total',legend=False,explode=[0,.2],autopct='%1.1f%%')
plt.ylabel('')
plt.title('Repeated Customer Percentage',fontsize=15)
```

### Conclution: We can see in the chart around 96.9% of the customers are not repeated only 3.1% customers are repeated.

# 9. How long people stay.

```python
df1 = data[data['is_canceled']==0]
```

```python
df1['total_stay'] = df1['stays_in_week_nights'] + df1['stays_in_weekend_nights']
```

```python
df2 = df1[['total_stay','hotel','is_canceled']]
```

```python
df3 = df2.groupby(['hotel','total_stay']).count().reset_index()
```

```python
dfCity = df3[df3['hotel']=='City Hotel']
```

```python
sns.barplot(data=dfCity[:20],x='total_stay',y='is_canceled').set(xlabel='No.of Days',ylabel='No. of Customer',title='Stay in City Hotel')
```

### Conclution: Above graph shows that the most of the customers Stays 2-3 days in City Hotels.

```python
dfResort = df3[df3['hotel']=='Resort Hotel']
```

```python
sns.barplot(data=dfResort[:20],x='total_stay',y='is_canceled',palette='Paired').set(xlabel='No.of Days',ylabel='No. of Customer',title='Days Stay in Resort Hotel')
```

### Conclution: Above graph shows that the most of the customers Stays around 1-7 days in Resort Hotels.
