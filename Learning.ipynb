# Learning Factor plot and Swarm plot etc.

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns

%matplotlib inline
```

## Data preparation for Factor Plot.

```python
exercise = sns.load_dataset('exercise')
```

```python
exercise.shape
```

```python
exercise.columns
```

```python
exercise.info()
```

```python
exercise.describe()
```

```python
exercise.isna().sum()
```

## 1. Factor Plot

```python
sns.factorplot(x='time', y='pulse', hue='kind', data=exercise,col='diet')
```

## 2. Joint plot.

```python
sns.jointplot(x='time',y='pulse',data=exercise)
```

## Data preparation for Swarm Plot.

```python
iris = sns.load_dataset('IRIS')
```

```python
iris.shape
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris.isna().sum()
```

## 3. Swarm Plot

```python
sns.swarmplot(x='species', y="petal_length", data=iris,linewidth=.2)
```

## 4. Pair Plot

```python
sns.pairplot(iris)
```
